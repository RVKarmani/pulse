## Optimized the prompt,
- Added a System Prompt for Guardrails (Reduce Hallucinations)
- Using reasoning models so that they can reason or think step by step which articles are relevant from the retrieved articles

```python
chat_completion_res = client.chat.completions.create(
    model=model,
    messages=[
        {
            "role": "system",
            "content": system_prompt,
        },
        {
            "role": "user",
            "content": user_prompt,
        }
    ],
    stream=stream,
    max_tokens=max_tokens,
    temperature=temperature,
    top_p=top_p,
    presence_penalty=presence_penalty,
    frequency_penalty=frequency_penalty,
    response_format=response_format,
    extra_body={
      "top_k": top_k,
      "repetition_penalty": repetition_penalty,
      "min_p": min_p
    }
)
```

```python
# Track what stage we're in
in_thinking = True
printed_ai_label = False

print("Thinking...\n")
# Stream tokens as they come
for chunk in chat_completion_res:
    # if 'choices' not in chunk:
    #     continue
    content = chunk.choices[0].delta.content

    if content:
        # Detect when thinking ends
        if in_thinking:
            print(content, end="", flush=True)
            if "</think>" in content:
                in_thinking = False
        else:
            if not printed_ai_label:
                print("\n\nAI Response :\n")  # move to next line after thinking block
                printed_ai_label = True
            print(content, end="", flush=True)

print("\n\nAI responses may include mistakes. Refer the citations given below.")
```
